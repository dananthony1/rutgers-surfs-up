```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

import datetime as dt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june10 = dt.date(2010, 6, 30)
may10 = dt.date(2010,5,31)
results10 = session.query(Measurement.tobs).\
filter(Measurement.date <= june10).\
filter(Measurement.date > may10).all()

june11 = dt.date(2011, 6, 30)
may11 = dt.date(2011,5,31)
results11 = session.query(Measurement.tobs).\
filter(Measurement.date <= june11).\
filter(Measurement.date > may11).all()

june12 = dt.date(2012, 6, 30)
may12 = dt.date(2012,5,31)
results12 = session.query(Measurement.tobs).\
filter(Measurement.date <= june12).\
filter(Measurement.date > may12).all()

june13 = dt.date(2013, 6, 30)
may13 = dt.date(2013,5,31)
results13 = session.query(Measurement.tobs).\
filter(Measurement.date <= june13).\
filter(Measurement.date > may13).all()

june14 = dt.date(2014, 6, 30)
may14 = dt.date(2014,5,31)
results14 = session.query(Measurement.tobs).\
filter(Measurement.date <= june14).\
filter(Measurement.date > may14).all()

june15 = dt.date(2015, 6, 30)
may15 = dt.date(2015,5,31)
results15 = session.query(Measurement.tobs).\
filter(Measurement.date <= june15).\
filter(Measurement.date > may15).all()

june16 = dt.date(2016, 6, 30)
may16 = dt.date(2016,5,31)
results16 = session.query(Measurement.tobs).\
filter(Measurement.date <= june16).\
filter(Measurement.date > may16).all()

june17 = dt.date(2017, 6, 30)
may17 = dt.date(2017,5,31)
results17 = session.query(Measurement.tobs).\
filter(Measurement.date <= june17).\
filter(Measurement.date > may17).all()

results = results10 + results11 + results12 + results13 + results14 + results15 + results16 + results17
print(results)
```

```python
#  3. Convert the June temperatures to a list.
temps = []
for temp in results:
    temps.append(temp[0])
print(temps)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
temps_df = pd.DataFrame(temps, columns=['June Temps'])
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
temps_df.describe()
```
